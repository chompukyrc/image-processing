```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
file_path = "data/duck1.jpg"
file_path_2 = "data/duck2.jpg"
```

# 1.2.1

```python
img = cv2.imread(file_path)

img1 = img.copy()
img2 = img.copy()
img3 = img.copy()
img4 = img.copy()

fig = plt.figure(figsize=(8, 8))

# RGB
img1 = img1[:,:,::-1]
r, g, b = cv2.split(img1)
fig.add_subplot(4, 4, 1)
plt.imshow(img1, cmap='gray')
fig.add_subplot(4, 4, 2)
plt.imshow(r, cmap='gray')
fig.add_subplot(4, 4, 3)
plt.imshow(g, cmap='gray')
fig.add_subplot(4, 4, 4)
plt.imshow(b, cmap='gray')

# HSV
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2HSV)
h,s,v = cv2.split(img2)
fig.add_subplot(4, 4, 5)
plt.imshow(img2, cmap='gray')
fig.add_subplot(4, 4, 6)
plt.imshow(h, cmap='gray')
fig.add_subplot(4, 4, 7)
plt.imshow(s, cmap='gray')
fig.add_subplot(4, 4, 8)
plt.imshow(v, cmap='gray')

# HSL
img3 = cv2.cvtColor(img3, cv2.COLOR_BGR2HLS)
h,l,s = cv2.split(img3)
fig.add_subplot(4, 4, 9)
plt.imshow(img3, cmap='gray')
fig.add_subplot(4, 4, 10)
plt.imshow(h, cmap='gray')
fig.add_subplot(4, 4, 11)
plt.imshow(l, cmap='gray')
fig.add_subplot(4, 4, 12)
plt.imshow(s, cmap='gray')

# YCrCb
img4 = cv2.cvtColor(img4, cv2.COLOR_BGR2YCR_CB)
Y,Cr,Cb = cv2.split(img4)
fig.add_subplot(4, 4, 13)
plt.imshow(img4, cmap='gray')
fig.add_subplot(4, 4, 14)
plt.imshow(Y, cmap='gray')
fig.add_subplot(4, 4, 15)
plt.imshow(Cr, cmap='gray')
fig.add_subplot(4, 4, 16)
plt.imshow(Cb, cmap='gray')
```

# 1.2.2

```python
img1 = cv2.imread(file_path)
img2 = cv2.imread(file_path_2)

print(img1.shape)
print(img2.shape)

fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter('video/output.mp4', fourcc, 5.0, (340, 340))

for i in range(30):
  dst = cv2.addWeighted(img1, 0.025 * i, img2, 1 - 0.025 * i, 0)
  out.write(dst)

for i in range(30):
  dst = cv2.addWeighted(img1, 1 - 0.025 * i, img2, 0.025 * i, 0)
  out.write(dst)

out.release()
```

```python
img = cv2.imread(file_path)
img = img[:,:,::-1]

print(img.shape[:2])

mask = np.zeros(img.shape[:2], dtype="uint8")
cv2.rectangle(mask, (100, 150), (250, 300), 255, -1)

masked = cv2.bitwise_and(img, img, mask=mask)
plt.imshow(masked)
```
