```python
import cv2
import numpy as np
import matplotlib.image as mpimg
import math

%matplotlib inline
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
file_path = "data/duck.jpg"
```

# 1.1.1

```python
img = cv2.imread(file_path)
img2 = img.copy()

fig = plt.figure(figsize=(12,12))

# BGR

b,g,r = cv2.split(img)
fig.add_subplot(1, 4, 1)
plt.imshow(img, cmap='gray')
fig.add_subplot(1, 4, 2)
plt.imshow(b, cmap='gray')
fig.add_subplot(1, 4, 3)
plt.imshow(g, cmap='gray')
fig.add_subplot(1, 4, 4)
plt.imshow(r, cmap='gray')

# bgr -> rgb

# RGB

img2 = img2[:,:,::-1]
r,g,b = cv2.split(img2)
fig.add_subplot(2, 4, 5)
plt.imshow(img2, cmap='gray')
fig.add_subplot(2, 4, 6)
plt.imshow(r, cmap='gray')
fig.add_subplot(2, 4, 7)
plt.imshow(g, cmap='gray')
fig.add_subplot(2, 4, 8)
plt.imshow(b, cmap='gray')
```

# 1.1.3

```python
img = cv2.imread(file_path)
img2=img.copy()
fig = plt.figure(figsize=(8,8))

#1
img = img[:,:,::-1]
img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.imshow(img, cmap="gray")

#2,3
shape =  (img.shape[0],img.shape[1])
img3=np.zeros(shape)
# print(img)
# print('*******************')
for k in range(img.shape[0]):
  for m in range(img.shape[1]):
        img3[k][m] = math.floor((img[k][m]/255)*(pow(2,4)-1))
# print(img3)
plt.imshow(img3 ,cmap='gray')
```

# 1.1.4

```python
img = cv2.imread(file_path)
img2=img.copy()
fig = plt.figure(figsize=(4, 4))

img = img[:,:,::-1]
img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.imshow(img, cmap="gray")

shape =  (img.shape[0],img.shape[1])
img3=np.zeros(shape)
# print(img)
# print('*******************')
for k in range(img.shape[0]):
  for m in range(img.shape[1]):
        img3[k][m] = math.floor((img[k][m]/255)*(pow(2,4)-1))
# print(img3)
plt.imshow(img3 ,cmap='gray')

img3 = cv2.resize(img3, (200, 200), interpolation = cv2.INTER_LINEAR)
x,y = np.mgrid[0:200,0:200]
fig3d = plt.figure()
ax = fig3d.add_subplot(projection='3d')
ax.plot_surface(x,y,img3,rstride=1,cstride=1,linewidth=0,cmap='gray')
```
